```python
import csv
import pandas as pd
data=pd.read_csv('E:\\Twitter paper\\DATASET_CREATED\\gephi\\Edge_file2.csv')
lan=['restaurant']
subset=data[data['Source'].isin(lan)]
subset.to_csv('E:\\Twitter paper\\DATASET_CREATED\\restaurant.csv',index = False)
print("done.....")
subset.shape

import matplotlib.pyplot as plt
data=pd.read_csv('E:\\Twitter paper\\DATASET_CREATED\\restaurant.csv')
a=0
b=0
c=0
lis=[]
for i in range (len(data)):
    if str(data['Label1'][i]) =='1':
        a=a+1
    elif str(data['Label1'][i]) == '0':
        b=b+1
    else:
        c=c+1
lis.append(a)
lis.append(b)
lis.append(c)
print(lis)
label=['positive','neutral','negative']
df={'Label':label,'Count':lis}
file= pd.DataFrame(df)
print(file)
file.plot.bar(x='Label')
```

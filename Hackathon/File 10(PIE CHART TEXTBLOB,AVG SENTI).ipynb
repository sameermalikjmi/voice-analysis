```python
import matplotlib.pyplot as plt
import pandas as pd
data =  pd.read_csv('E:\\Twitter paper\\DATASET_CREATED\\Zomato_ENGLISH.csv')
#data =  pd.read_csv('E:\\Zomato_ENGLISH.csv')
a=0
b=0
c=0
lis=[]
for i in range (len(data)):
    if str(data['sentiments'][i]).replace("[",'').replace("]","") =='1': #positive (a)
        a=a+1
    elif str(data['sentiments'][i]).replace("[",'').replace("]","") =='0': #neutral(b)
        b=b+1
    else:
        c=c+1 #negative(c) 
lis.append(a)
lis.append(b)
lis.append(c)
print(lis)
label=['Positive','Neutral','Negative']
file={'Label':label,'COUNT':lis}
df=pd.DataFrame(file)
print(df)
df.to_csv("E:\\Twitter paper\\DATASET_CREATED\\sentimnet_count.csv", index=False)
#df.to_csv("E:\\sentimnet_count.csv", index=False)
colors = ["#2ca02c","#ff7f0e", "#1f77b4"]
#plt.pie(medal_data, colors=colors,autopct='%1.1f%%')
plt.pie(lis, labels=label, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("overall sentiment pie chart\n"+"textblob")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
#data =  pd.read_csv('E:\\Twitter paper\\DATASET_CREATED\\Zomato_ENGLISH.csv')
data =  pd.read_csv('E:\\Zomato_ENGLISH.csv')
a=0
b=0
c=0
lis=[]
for i in range (len(data)):
    if float(str(data['average sentiment'][i].replace("[",'').replace("]",""))) > 0:
        a=a+1
    elif float(str(data['average sentiment'][i].replace("[",'').replace("]",""))) < 0:
        b=b+1
    else:
        c=c+1
lis.append(a)
lis.append(b)
lis.append(c)
print(lis)
label=['Positive','Negative','Neutral']
file={'Label':label,'COUNT':lis}
df=pd.DataFrame(file)
print(df)
df.to_csv("E:\\Twitter paper\\avg_sentimnet_count.csv", index=False)
#df.to_csv("E:\\avg_sentimnet_count.csv", index=False)
colors = ["#2ca02c", "#1f77b4","#ff7f0e"]
#plt.pie(medal_data, colors=colors,autopct='%1.1f%%')
plt.pie(lis, labels=label, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("overall sentiment pie chart\n"+"using aspect sentimnet")
plt.show()
```
